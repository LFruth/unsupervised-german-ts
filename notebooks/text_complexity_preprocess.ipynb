**Merge articles**

```python
# combine sentences to original article + simplifications
parallel_corpus = parallel_corpus.drop(columns=['Rating', 'Sentence_Id'])
parallel_corpus['Original_Sentence'] = parallel_corpus.groupby(['Article_ID'])['Original_Sentence'].transform(lambda x : ' '.join(x))
parallel_corpus['Simplification'] = parallel_corpus.groupby(['Article_ID'])['Simplification'].transform(lambda x : ' '.join(x))
parallel_corpus = parallel_corpus.drop_duplicates()
```

```python
source = source.drop(columns=['Sentence ID', 'Number of Words'])
source['Sentence'] = source.groupby(['Article ID'])['Sentence'].transform(lambda x : ' '.join(x))
source = source.drop_duplicates()
```
